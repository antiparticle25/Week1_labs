```python
# concat dataFrames
final_data_train = pd.concat([X_train2,X_train1,x_train_num_df],axis=1)
final_data_test = pd.concat([X_test2,X_test1,x_test_num_df],axis=1)
final_data_train.head()
```

```python
#import libraries
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
import math
```

```python
#Training the linear model
lm = linear_model.LinearRegression()
#"train" our linear model based on the data available on X_train and y_train
model = lm.fit(final_data_train,y_train)
```

```python
# We make predictions on X_test
predictions  = lm.predict(final_data_test)
predictions
```

```python
#final_data_test.isna().sum()
```

```python
#Apply linear regression
print("The R2 of the linear model is: ",round(r2_score(y_test, predictions),2))
```

```python
#Computing MSE
mse = mean_squared_error(y_test, predictions)
print(round(mse,2))
```

```python
#Computing RMSE
rmse = math.sqrt(mse)
print(round(rmse,2))
```

```python
#Computing MAE
mae = mean_absolute_error(y_test, predictions)
print(round(mae,2))
```
